## Web Map GeoJSON Export

````python
geojson_path = DATA_OUT / 'schools_webmap.geojson'
if geojson_path.exists():
    with open(geojson_path) as f:
        gj = json.load(f)
    n = len(gj['features'])
    size_kb = geojson_path.stat().st_size // 1024
    print(f'GeoJSON ready: {n:,} features, {size_kb:,} KB')
    print('Sample properties:', list(gj['features'][0]['properties'].keys()))
else:
    print('Run run_phase4_static() first')
````
